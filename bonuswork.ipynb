```python
import os
import csv
import pandas as pd
import matplotlib as plt

credits_file = 'tmdb_5000_credits.csv'
movies_file = 'tmdb_5000_movies.csv'

credits_data = pd.read_csv(credits_file)
movies_data = pd.read_csv(movies_file)
```

```python
credits_data["Top 5 Cast"]=""
i=-1
#credits_data=next(credits_data)
for index,row in credits_data.iterrows():
    name_data=[]
    cast_key= str(row[2]).split(':')

    search="name"
    for key in cast_key:
        if search in key:
            name_data.append(cast_key[cast_key.index(key)+1])
            print(name_data)
        for name in name_data:
            names=[]
            name=str(name).split(",") 
            names.append(name[0])
            names_5=list(names)
    print(names_5)
    i=i+1
    credits_data["Top 5 Movie Cast"][index]=names_5[:5]
    
```
